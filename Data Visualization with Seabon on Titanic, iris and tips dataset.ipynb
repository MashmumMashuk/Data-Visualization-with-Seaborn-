We have used **tips**, **iris** and another very famous dataset **titanic** for these exercises. These datasets are included in the seaborn's built-in dataset library.<br>
Let's do some exploratory data analysis using seaborn and present useful plots for data visualization!<br>

```python
#Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```



* set style to '`white`'
* load titanic dataset into '`titanic`', tips into '`tips`' and iris into '`iris`'

```python
#Setting the style to white
sns.set_style('white')
```

```python
#Loading the respective datasets
titanic= sns.load_dataset('titanic')
tips= sns.load_dataset('tips')
iris= sns.load_dataset('iris')
```

**Tips dataset** 

```python
#Boxplot on tips earned on each day

fig=plt.figure(figsize=(4.5,3))

sns.boxplot(x='day', y='tip', data=tips, hue='sex')
```

**Tips dataset**

```python
# jointplot on the tips earned based on total bill

sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', height =5)
```

**Titanic Dataset.** 

```python
#Distribution plotting on fare of titanic
fig=plt.figure(figsize=(4.5,3))

sns.distplot(x=titanic['fare'], kde=False, bins = 25, color='green')
plt.xlabel('fare')
```

**Titanic Dataset** 

```python
#Boxplot on titanic dataset illustrating the passengers age in specific class of the cruise.

fig=plt.figure(figsize=(6,4))

sns.boxplot(x='class', y='age', data=titanic, hue='alive')

plt.tight_layout()
```

**Titanic Dataset** 

```python
# JointPlot  on fare and age of the passengers in titanic


sns.jointplot(x='fare', y='age', data=titanic)
```

**plotting how many male and female were on titanic. Separated based on their class.**

```python
#Count Plot on the gender of passengers (male/female) present on titanic

fig=plt.figure(figsize=(4.5,3))

sns.countplot(x='sex', data=titanic, hue='class')
```

**Titanic Dataset**


```python
# Heat map : Illustrating the correlation of each data with another in the titanic dataset

fig=plt.figure(figsize=(6,4))

sns.heatmap(data=titanic.corr(), annot=titanic.corr(), cmap='coolwarm', vmin=-1 ,vmax=1)

plt.tight_layout()
```
